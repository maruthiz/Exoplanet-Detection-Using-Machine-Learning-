```python
# Define a mapping dictionary to replace labels
label_mapping = {1: 0, 2: 1}

# Use the map function to replace labels
train_df['LABEL'] = train_df['LABEL'].map(label_mapping)

print("Lets replace labels!")

# Check the unique classes after replacement
unique_classes = train_df['LABEL'].unique()
print(f"There are {len(unique_classes)} classes in the data:-")
print(f"{unique_classes[0]} - Stars with Exoplanets\n{unique_classes[1]} - Stars without Exoplanets")
```

```python
# Drop the 'LABEL' column to plot only the flux values
plot_df = train_df.drop(columns='LABEL')

# Extract time data by removing 'FLUX.' from column names
time = [int(col.replace('FLUX.', '')) for col in plot_df.columns]

# Function to plot flux variation of a star
def plot_flux_variation(df, candidate, exo=True):
    color = 'b' if exo else 'm'
    plt.figure(figsize=(15, 5))
    plt.plot(time, df.iloc[candidate - 1], linewidth=0.5, color=color)
    title = f"Flux Variation of star {candidate} {'with' if exo else 'without'} Exoplanets"
    plt.title(title, color='olive' if exo else 'tab:green')
    plt.xlabel("Time")
    plt.ylabel("Flux Variation")

exo, n_exo = [4, 14, 34], [99, 199, 2999]

for candidate in exo + n_exo:
    is_exo = candidate in exo
    plot_flux_variation(plot_df, candidate, exo=is_exo)
```

```python


# Create subplots for box plots
fig, axes = plt.subplots(1, 4, figsize=(20, 5))

# Set the overall title
fig.suptitle("Box Plot to Visualize Outliers", ha='right', color='red', weight='bold')

for i in range(1, 4):
    # Create a box plot for each flux value within each class
    sns.boxplot(x='LABEL', y='FLUX.' + str(i), data=train_df, ax=axes[i - 1])
    
    # Set titles and labels
    axes[i - 1].set_title("FLUX " + str(i) + "\n", color='b', fontsize=13)
    axes[i - 1].set_xlabel("")
    axes[i - 1].set_ylabel("")

# Adjust spacing between subplots
plt.tight_layout()

# Show the plots
plt.show()
```

```python
# Notice the extreme outliers
extreme_outliers = train_df[train_df['FLUX.2'] > 0.25e6]
extreme_outliers
```

```python
# Calculate the mean and standard deviation of FLUX.2
mean_flux2 = np.mean(train_df['FLUX.2'])
std_flux2 = np.std(train_df['FLUX.2'])

# Define a threshold for extreme outliers (e.g., 3 standard deviations away from the mean)
threshold = 3 * std_flux2

# Identify extreme outliers
extreme_outliers = train_df[train_df['FLUX.2'] > mean_flux2 + threshold]

# Display the extreme outliers
print("Extreme Outliers:")
print(extreme_outliers)
```

```python
# Drop the extreme outlier

train_df.drop(extreme_outliers.index, axis = 0, inplace = True)  
print("Dropped Outliers!")
```

```python
# Extract dependent and independent features
x = train_df.drop(['LABEL'], axis = 1)
y = train_df.LABEL

print(f"\nX train array:-\n{x.values}\n\nY train array:-\n{y.values}")
```

```python
# Splitting this dataset into training and testing set
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 0)
```

```python
# Feature scaling
from sklearn.preprocessing import StandardScaler 

sc = StandardScaler()
X_train_sc = sc.fit_transform(X_train)
X_test_sc = sc.transform(X_test)

# Checking the minimum, mean and maxmum value after scaling
print("X_train after scaling ~\n")
print(f"Minimum:- {round(np.min(X_train_sc),2)}\nMean:- {round(np.mean(X_train_sc),2)}\nMax:- {round(np.max(X_train_sc), 2)}\n")
print("--------------------------------\n")
print("X_test after scaling ~\n")
print(f"Minimum:- {round(np.min(X_test_sc),2)}\nMean:- {round(np.mean(X_test_sc),2)}\nMax:- {round(np.max(X_test_sc), 2)}\n")
```

```python
#Feature Scaling 
mean_train = np.mean(X_train, axis=0)
std_train = np.std(X_train, axis=0)

# Scale the training and test data
X_train_sc = (X_train - mean_train) / std_train
X_test_sc = (X_test - mean_train) / std_train

# Check the minimum, mean, and maximum value after scaling
print("X_train after scaling ~\n")
print(f"Minimum:- {round(np.min(X_train_sc), 2)}\nMean:- {round(np.mean(X_train_sc), 2)}\nMax:- {round(np.max(X_train_sc), 2)}\n")
print("--------------------------------\n")
print("X_test after scaling ~\n")
print(f"Minimum:- {round(np.min(X_test_sc), 2)}\nMean:- {round(np.mean(X_test_sc), 2)}\nMax:- {round(np.max(X_test_sc), 2)}\n")
```

```python
# Fiting the KNN Classifier Model on to the training data
from sklearn.neighbors import KNeighborsClassifier as KNC

# Choosing K = 10
knn_classifier = KNC(n_neighbors=5,metric='minkowski',p=2)  
'''metric is to be by default minkowski for p = 2 to calculate the Eucledian distances'''

# Fit the model
knn_classifier.fit(X_train_sc, y_train)

# Predict
y_pred = knn_classifier.predict(X_test_sc)

# Results
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report, roc_curve, auc

print('\nValidation accuracy of KNN is', accuracy_score(y_test,y_pred))
print("\n-------------------------------------------------------")
print ("\nClassification report :\n",(classification_report(y_test,y_pred)))

#Confusion matrix
plt.figure(figsize=(15,11))
plt.subplots_adjust(wspace = 0.3)
plt.suptitle("KNN Performance before handling the imbalance in the data", color = 'r', weight = 'bold')
plt.subplot(221)
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True,cmap="Set2",fmt = "d",linewidths=3, cbar = False,
           xticklabels=['nexo', 'exo'], yticklabels=['nexo','exo'], square = True)
plt.xlabel("True Labels", fontsize = 15, weight = 'bold', color = 'tab:pink')
plt.ylabel("Predicited Labels", fontsize = 15, weight = 'bold', color = 'tab:pink')
plt.title("CONFUSION MATRIX",fontsize=20, color = 'm')

#ROC curve and Area under the curve plotting
predicting_probabilites = knn_classifier.predict_proba(X_test_sc)[:,1]
fpr,tpr,thresholds = roc_curve(y_test,predicting_probabilites)
plt.subplot(222)
plt.plot(fpr,tpr,label = ("AUC :",auc(fpr,tpr)),color = "g")
plt.plot([1,0],[1,0],"k--")
plt.legend()
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title("ROC - CURVE & AREA UNDER CURVE",fontsize=20, color = 'm')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_curve, auc
import matplotlib.pyplot as plt
import seaborn as sns

# Split the data into a training set and a testing set
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Initialize the Random Forest Classifier
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)

# Fit the model on the training data
rf_classifier.fit(X_train, y_train)

# Make predictions on the testing data
y_pred = rf_classifier.predict(X_test)

# Calculate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy * 100:.2f}%')

# Generate a confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)

# Print a classification report
class_report = classification_report(y_test, y_pred, target_names=['Stars without Exoplanets', 'Stars with Exoplanets'])

# Plot confusion matrix
plt.figure(figsize=(10, 8))
plt.suptitle("Random Forest Classifier Performance", color='r', weight='bold')

plt.subplot(221)
sns.heatmap(conf_matrix, annot=True, cmap="Set2", fmt="d", linewidths=3, cbar=False,
            xticklabels=['nexo', 'exo'], yticklabels=['nexo', 'exo'], square=True)
plt.xlabel("True Labels", fontsize=15, weight='bold', color='tab:pink')
plt.ylabel("Predicted Labels", fontsize=15, weight='bold', color='tab:pink')
plt.title("CONFUSION MATRIX", fontsize=20, color='m')

# ROC curve and Area under the curve plotting
predicting_probabilities = rf_classifier.predict_proba(X_test)[:, 1]
fpr, tpr, thresholds = roc_curve(y_test, predicting_probabilities, pos_label=1)
plt.subplot(222)
plt.plot(fpr, tpr, label=f"AUC: {auc(fpr, tpr):.2f}", color="g")
plt.plot([0, 1], [0, 1], "k--")
plt.legend()
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title("ROC CURVE & AREA UNDER CURVE", fontsize=20, color='m')


plt.show()
```
